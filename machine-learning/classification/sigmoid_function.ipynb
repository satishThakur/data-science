```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    """
       Here x is one dimentional array and the function returns the value of 
       sigmoid function at each x. 
    """
    
    return 1 / (1 + np.exp(-x))
    
```

```python
x = np.arange(-10,10, 1)
sig_x = sigmoid(x)

fig,ax = plt.subplots(1, 1, constrained_layout=True, figsize=(8, 4))
ax.plot(x, sig_x, c="b", label="sigmoid")
ax.set_xlabel("X")
ax.set_ylabel("Sigmoid(x)")
plt.show()
```
