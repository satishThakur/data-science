```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.arange(0.01, 1.0, .05)
print(z)
```

```python
l = -np.log(z)
l1 = -np.log(1-z)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,6))

ax.plot(z,l, c="r", label="y=1")
ax.plot(z,l1, c="b", label="y=0")
ax.set_ylabel('Z')
ax.set_xlabel('Loss')
plt.legend()
plt.draw()
```
