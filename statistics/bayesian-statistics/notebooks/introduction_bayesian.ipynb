## Probability, Statistics and Inference
In probability theory we generally have probability of a event or events and from that we calculate probability of other derivations. For example given probability of coin toss being head is 0.5, what is probability of getting 4 heads in 10 tosses etc.
But in real life we have observations and no probability - that is where Statistics comes in. We use statistical methods to derive probability of certain events. Hence in short statistics is probability in reverse.
Once we have derived the probability we can make inferences about the data or certain aspects of the data. In this context finding the probability from data is called inference.
