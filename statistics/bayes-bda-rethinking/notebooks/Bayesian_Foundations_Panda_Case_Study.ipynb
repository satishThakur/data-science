## Probability (Bayesian View)
Probability represents uncertainty in parameters, not randomness in the world.

Sources:
- Measurement error
- Model uncertainty

Laplace–Jeffreys–Cox–Jaynes interpretation.

## Conjectures
Multiple conjectures exist simultaneously and compete at the same time.

A conjecture that can produce data in more ways is more plausible.

## Core Objects
- Parameter: each discrete value is a conjecture
- Likelihood: compatibility of data with conjecture
- Prior: plausibility before data
- Posterior: updated plausibility after data

## Bayesian Update (Discrete)
$P(\theta_i|y)=\frac{P(y|\theta_i)P(\theta_i)}{\sum_j P(y|\theta_j)P(\theta_j)}$

Denominator is expected likelihood.

## Prediction
Prediction = averaging likelihood over parameter distribution.

## Panda Case Study
Two species SA and SB, equally common.

$P(T|SA)=0.1$, $P(T|SB)=0.2$

## Prior
$P(SA)=0.5$, $P(SB)=0.5$

## Observation: Twin Birth
$P(SA|T)=1/3$, $P(SB|T)=2/3$

## Posterior Predictive
$P(T_{next}|T)=0.165$

## Genetic Test
$P(A^+|SA)=0.8$, $P(A^+|SB)=0.35$

## Updated Posterior
$P(SA|A^+)=0.533$, $P(SB|A^+)=0.467$

## Final Prediction
$P(T_{next}|T,A^+)=0.1467$

## Takeaways
Beliefs update, predictions follow.
